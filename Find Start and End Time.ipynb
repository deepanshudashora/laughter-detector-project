```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
import IPython.display as ipd
import soundfile as sf
import sounddevice as sd
```

```python
audio_path = '8.mp3'
x , sr = librosa.load(audio_path, sr=None)#sr=None means the signal is sampled at original sample rate
print('Sampling Rate = ', sr/1000, 'kHz')
print('Signal Length = ', x.shape[0]/sr, 'seconds')
```

```python
#enter time from start to end
t_start = 327.5
t_end = 32
clip = x[int(t_start*sr):int(t_end*sr)]

#display waveform
%matplotlib inline
plt.figure(figsize=(14, 5))
librosa.display.waveplot(clip, sr=sr)
sd.play(clip)#Put headphones on
```
